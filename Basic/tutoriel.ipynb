```python

# https://www.youtube.com/watch?v=i-xbj0owgEE
# Plotly 한번에 끝내기 - 모던한 인터랙티브 시각화 라이브러리
# 이수한컴퓨터연구소

import numpy as np
import pandas as pd
from urllib.request import urlopen
import json
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
from plotly.subplots import make_subplots
from plotly.validators.scatter.marker import SymbolsrcValidator
```

<h1> Scatter Plot

```python
print(iris)
```

```python
fig = px.scatter(iris, x='petal_width', y='petal_length')
fig.show()
```

```python
tips = px.data.tips()
print(tips)
# facet raw는 day
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', facet_row='day')
fig.show()
fig = px.scatter(tips, x='total_bill', y='tip', color='sex', facet_col='day')
fig.show()
```

```python
t = np.linspace(0, 10, 100)
y1 = np.random.randn(100).cumsum()
y2 = np.random.randn(100).cumsum()

fig = go.Figure()
fig.add_trace(go.Scatter(x=t, y=y1, name='red', mode='markers', marker_color='darkred'))
fig.add_trace(go.Scatter(x=t, y=y2, name='blue', mode='markers', marker_color='darkblue'))
fig.update_traces(mode='markers', marker_line_width=1, marker_size=10)
fig.show()
```

```python
fig = go.Figure(data=go.Scatter(x=[1,2,3,4,5], y=[11,12,13,14,15], mode='markers', marker=dict(size=[20,40,60,80,100], color=[1,2,3,4,5])))
fig.show()
```

```python
gap_minder = px.data.gapminder()
gap_minder_2007 = gap_minder.query("year == 2007")
gap_minder_2007
```

```python
# trendline option을 통해 triendline을 알 수 있음
# size를 string으로 지정하면 해당 변수로 marker의 크기를 조정할 수 있음
fig = px.scatter(gap_minder_2007, x='gdpPercap', y="lifeExp", size='pop', color='continent', hover_name='country',
                 log_x=True, size_max=60, trendline='lowess')
fig.show()
```

```python
# legend의 y를 통해 위치 조정, x로도 위치조정 traceorder 순서대로 출력

x = np.array([1,2,3,4,5])
y = np.array([1,3,2,3,1])

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, name='linear', line_shape='linear'))
fig.add_trace(go.Scatter(x=x, y=y+5, name='spline', line_shape='spline', hoverinfo="text+name"))
fig.add_trace(go.Scatter(x=x, y=y+10, name='vhv', line_shape='vhv', hoverinfo="text+name"))
fig.add_trace(go.Scatter(x=x, y=y+15, name='hvh', line_shape='hvh', hoverinfo="text+name"))
fig.add_trace(go.Scatter(x=x, y=y+20, name='vh', line_shape='vh', hoverinfo="text+name"))
fig.update_traces(hoverinfo='text+name', mode='lines+markers')
fig.add_trace(go.Scatter(x=x, y=y+25, name='hv', line_shape='hv'))
fig.update_layout(legend=dict(y=0.5, font_size=16))
fig.show()
fig.update_layout(legend=dict(y=0, traceorder='reversed', font_size=16))
fig.show()
```

```python
N = 100
x = np.linspace(0, 1, N)
y0 = np.random.rand(N) + 5
y1 = np.random.rand(N)
y2 = np.random.rand(N) -5

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y0, mode='markers', name='markers'))
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines+markers', name='markers + linemarkers'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='linemarkers'))
fig.show()
```

```python
apple = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
print(apple)

fig = px.line(apple, x='Date', y='AAPL.Close', range_x=['2016-02-01','2017-03-01'])
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
fig = go.Figure(data = [go.Candlestick(x=apple['Date'],
                                       open=apple['AAPL.Open'],
                                       high=apple['AAPL.High'],
                                       low=apple['AAPL.Low'],
                                       close=apple['AAPL.Close'])])
fig.show()


fig = go.Figure(data = [go.Candlestick(x=apple['Date'],
                                       open=apple['AAPL.Open'],
                                       high=apple['AAPL.High'],
                                       low=apple['AAPL.Low'],
                                       close=apple['AAPL.Close'],
                                       increasing_line_color='red',
                                       decreasing_line_color='blue')])
fig.show()


```
